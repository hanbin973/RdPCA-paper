```r
library(Seurat)
library(scran)
```

```r
obj <- readRDS('../../data/pbmc_10k_v3/pbmc_10k_v3.rds')
```

```r
DimPlot(obj)
```

```r
DimPlot(obj, reduction='tsne')
```

```r
obj <- RunTSNE(obj, dims=1:50)
```

```r
obj_subset <- subset(obj, idents=c('CD14+ Monocytes', 'Dendritic cell', 'pre-B cell'))
```

```r
chname <- c('CD14+ Monocytes'='Monocytes', 
    'Dendritic cell'='DCs',
    'pre-B cell'='B cells')
newident <- chname[as.vector(Idents(obj_subset))]
names(newident) <- names(Idents(obj_subset))
Idents(obj_subset) <- newident
```

```r
Idents(obj_subset)
```

```r
# subset relavent genes
mat_subset <- t(as.data.frame(obj_subset$RNA@counts[c('CD14', 'LYZ', 'CD79A', 'CD79B', 'MS4A1'),]))
```

# scran size factor

```r
sf <- calculateSumFactors(obj_subset$RNA@counts)
```

```r
mat_scran <- data.frame(mat_subset/sf)
mat_scran$method <- 'scran'
```

# sctransform

```r
obj_sct <- SCTransform(obj_subset)
```

```r
mat_sct <- as.data.frame(t(obj_sct$SCT@counts[c('CD14', 'LYZ', 'CD79A', 'CD79B', 'MS4A1'),]))
mat_sct$method <- 'SCT'
```

```r
mat_orig <- as.data.frame(t(obj_sct$RNA@counts[c('CD14', 'LYZ', 'CD79A', 'CD79B', 'MS4A1'),]))
mat_orig$method <- 'Raw'
```

```r
celltype <- data.frame(celltype=Idents(obj_subset))
df_fig1e <- rbind(cbind(celltype, mat_orig), cbind(celltype, mat_scran), cbind(celltype, mat_sct))
```

```r
write.csv(df_fig1e, 'fig1e.csv')
```
